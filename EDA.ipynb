```python
df['Month'].unique()
```

```python
df['Month']=df['Month'].map({
    1:'JAN',
    2:'FEB',
    3:'MAR',
    4:'APR',
    5:'MAY',
    6:'JUN',
    7:'JUL',
    8:'AUG',
    9:'SEP',
    10:'OCT',
    11:'NOV',
    12:'DEC'
})
```

```python
plt.figure(figsize=(12,6))
sns.countplot(df['Month'],palette='Set2')
plt.title('Flights per Month',size=20)
#plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(df['Month'], df['Price'], palette='Set2')
plt.title('Month vs Price', size=20)
#plt.xticks(rotation=90)
```

```python
df['Duration_Hr']
```

```python
df['Duration_Hr'],_=df['Duration_Hr'].str.split(':',1).str
```

```python
df['Duration_Hr']
```

```python
df['Duration_Hr'].dtype
```

```python
df['Duration_Hr']=df['Duration_Hr'].astype(int)
```

```python
df['Duration_Hr'].dtype
```

```python
df['Duration_bool']=(df['Duration_Hr']*60)+df['Duration_Min']
```

```python
df['Duration_bool']
```

```python
plt.figure(figsize=(12,6))
sns.scatterplot(df['Duration_bool'], df['Price'], palette='Set2')
plt.title('Duration vs Price', size=20)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(df['Total_Stops'], df['Price'], palette='rainbow')
plt.title('Stops vs Price', size=20)
```

```python
plt.figure(figsize=(12,6))
sns.barplot(df['Additional_Info'], df['Price'], palette='Set1')
plt.title('Additional Info vs Price', size=20)
plt.xticks(rotation=90)
```
